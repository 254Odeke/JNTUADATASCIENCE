### Matplolib:

* pyplot Module is used to represent the data in the graphical format.
* It is a cross platform for making 2d plots from the given data.

**Different ploting methods**:
* line plot
* Bar graph

```python
### importing matplotlib:

import matplotlib.pyplot as plt
```

### Line plot:

* Line plot is the linear grap,in which all points are join by a line.
* plot()

```python
x = [1,2,3,4,5]
y = [1,2,3,4,5]

plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,'g')
plt.show()
```

```python
plt.plot(x,y,color = '0')
```

```python
### plot with xlabel and ylabel
y = [i**2 for i in x]
plt.plot(x,y,c = 'y')
plt.title("Graph of x values and power of x values")
plt.xlabel("x-value")
plt.ylabel("power of X values")
plt.show()
```

```python
import numpy as np
x = np.linspace(0,4,30)
y = np.logspace(0,4,30)
print(x)
```

```python
plt.plot(x,y,color = 'r',label = "linear graph")
plt.plot(x**2,y*3,color = 'g',label = "squares")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("line plot")
plt.legend(loc = "center")
plt.show()
```

```python
plt.plot(x,y,color = 'r',label = "linear graph")
plt.plot(x**2,y*3,color = 'g',label = "squares")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("line plot")
plt.legend(loc = "upper center")
plt.show()
```

```python
help(plt.plot)
```

```python
plt.plot(x,y,color = 'r',label = "linear graph",linestyle = '--')
plt.plot(x**2,y*3,color = 'g',label = "squares",linestyle = '-.')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("line plot")
plt.legend(loc = "upper center")
plt.show()
```

```python
plt.plot(x,y,'#000000',linestyle = "-.",linewidth = 1,marker = 'P',label = "line1")
plt.plot(x+2,y,'#0000ff',linestyle = '--',marker = 'P',label = 'line2')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Multiple plots")
plt.legend(loc = "best")
plt.show()
```

**Subplots**:


```python
fig,a = plt.subplots(2,2)
a[0][0].plot(x,x*x)
a[0][0].set_title('square')
a[0][1].plot(x,np.sqrt(x))
a[0][1].set_title('square root')
a[1][0].plot(x,np.exp(x))
a[1][0].set_title('Exponential')
a[1][1].plot(x,np.log(x))
a[1][1].set_title('Logarithmic')
plt.show()
```

```python
a1 = plt.subplot2grid((3,3),(0,0),colspan = 2)
a2 = plt.subplot2grid((3,3),(0,2),rowspan = 3)
a3 = plt.subplot2grid((3,3),(1,0),rowspan = 2,colspan = 2)
a2.plot(x,x*x)
a2.set_title('square')
a3.plot(x,np.exp(x))
a3.set_title('exponential')
a1.plot(x,y)
a1.set_title('sample')
plt.show()
```

**Boxplot**:
* This is also called as whisker plot
* box()
* main terms in box plot
    * maximum
    * minimum
    * first quartile
    * median
    * third quartile
    
 <img src = "https://www.onlinemathlearning.com/image-files/box-plot.png" height = "500" width = "500">

```python
data = np.random.normal(100,20,200)
data
```

```python
plt.boxplot(data)
plt.show()
```

```python
data.max()
```

```python
data.min()
```

```python
np.median(data)
```

```python
a = np.random.randint(50,100,10)
b = np.random.randint(40,90,10)
c = np.random.randint(30,100,10)
plt.boxplot([a,b,c],labels = ["box1","box2","box3"])
plt.show()
```

**Area plot**:

* stackplot()

```python
days = [1,2,3,4]
working = [8,11,9,10]
sleeping = [8,10,11,12]
plt.stackplot(days,working,sleeping,colors = ['r','g'],labels = ['work','sleep'])
plt.legend()
plt.show()
```

**pie charts**:
* pie()

```python
subjects = ['Python','java','C']
marks = [90,80,50]
plt.pie(marks,labels = subjects,colors = ['r','g','m'],radius = 1)
plt.show()
```

```python
subjects = ['Python','java','C']
marks = [90,80,50]
plt.pie(marks,labels = subjects,colors = ['r','g','m'],radius = 1,explode = [0,0,1])
plt.show()
```

```python
subjects = ['Python','java','C']
marks = [90,80,50]
plt.pie(marks,labels = subjects,colors = ['r','g','m'],radius = 1,explode = [1,1,1])
plt.show()
```

```python
subjects = ['Python','java','C']
marks = [90,80,50]
plt.pie(marks,labels = subjects,colors = ['r','g','m'],radius = 1,explode = [0,1,0])
plt.show()
```
