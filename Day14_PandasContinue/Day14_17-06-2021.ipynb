## Merging

```python
d1 = {
    "Name":["chinnu","chinni","chintu","chinna"],
    "Fruit": ["orange","apple","graps","mango"],
    "color" : ["black","blue","white","Pink"]
}
```

```python
import numpy as np
import pandas as pd
```

```python
df1 = pd.DataFrame(d1)
df1
```

```python
d2 = {
    "Name":["chinnu","chinni","chintu","chinna"],
    "Fruit": ["orange","apple","graps","mango"],
    "age" : [23,34,23,21]
}
```

```python
df2 = pd.DataFrame(d2)
df2
```

```python
df1.append(df2)
```

```python
pd.concat([df1,df2],axis = 1)
# axis = 1 - columns , axis = 0 - rows 
```

```python
df3 = pd.merge(df1,df2)
```

# Grouping

```python
data[data["type"] == "Iris-setosa"] # data masking
```

```python
g_data.last()
```

```python
g_data.get_group("Iris-virginica").min()
```

```python
g_data.get_group("Iris-virginica").mean()
```
