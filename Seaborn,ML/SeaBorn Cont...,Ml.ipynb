```python
import seaborn as sns
```

**Joint Plot:**
* Combines two different plots into one.

```python
len(sns.get_dataset_names())
```

```python
sns.get_dataset_names()
```

```python
iris = sns.load_dataset("iris")
iris.head()
```

```python
iris.columns
```

```python
sns.jointplot(x = "sepal_length",y = "sepal_width",data = iris,kind = "reg" )
```

```python
import matplotlib.pyplot as plt
sns.jointplot(x = "petal_length", y = "petal_width",data = iris,kind = "reg")
plt.show()
```

```python
sns.jointplot(x = "petal_length",y = "petal_width",data = iris)
plt.show()
```

```python
help(sns.jointplot)
```

```python
sns.jointplot(x = "petal_length",y = "petal_width",data = iris,kind = "kde",color = "m")
plt.show()
```

```python
sns.jointplot(x = "petal_length",y = "petal_width",data = iris,kind = "resid",color = "g")
plt.show()
```

```python
sns.jointplot(x = "petal_length",y = "petal_width",data = iris,kind = "hex",color = "g")
plt.show()
```

```python
sns.jointplot(x = "petal_length",y = "petal_width",data = iris,color = 'k').plot_joint(sns.kdeplot,n_levels = 10)
plt.show()
```

```python
import numpy as np
x,y = np.random.randn(2,300)
sns.jointplot(x,y,kind = "hex").set_axis_labels("x","y")
plt.show()
```

**Pairplot**:

```python
sns.set_style("ticks")
sns.pairplot(iris,hue = "species",diag_kind = "kde",kind = "scatter",palette = "husl")
plt.show()
```

```python
sns.pairplot(iris,hue = "species",markers = ["o","s",'d'])
plt.show()
```

**Strip plot**

```python
sns.stripplot(x = "species",y = "sepal_length",data = iris)
plt.show()
```

```python
sns.stripplot(x = "species",y = "petal_length",data = iris)
plt.show()
```

**HeatMap**

```python
a = np.random.randn(10,20)
sns.heatmap(a)
plt.show()
```

```python
sns.heatmap(a,vmin = 0,vmax = -5)
plt.show()
```

```python
plt.figure(figsize = (10,5))
sns.heatmap(a,annot = True)
plt.show()
```

```python
plt.figure(figsize = (10,5))
sns.heatmap(a,annot = True,linewidth = 0.5)
plt.show()
```

```python
sns.heatmap(a,cmap = "YlGnBu")
plt.show()
```
